```python
#a) hiptesis de causalidad: Redacta una hiptesis sobre la causalidad entre las dos variables, a su vez, establece la variable dependiente y la variable independiente.
#Hiptesis de casualidad: Un mayor porcentaje de poblacin con trastornos depresivos est asociado con un mayor porcentaje de poblacin con
#trastornos de ansiedad.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/PaulinaAragon/ESTADISTICA/refs/heads/main/mental_health.csv')
df.head(50)
# eliminar registros  con valores faltantes
df.dropna (inplace=True)
X = df["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"] # Variable independiente
Y = df["Depressive disorders (share of population) - Sex: Both - Age: Age-standardized"] # Variable dependiente

#b) recta de regresin ajustada: Realiza los clculos necesarios para encontrar la recta de regresin ajustada. Incluye los procedimientos que te llevaron a la recta de regresin
import matplotlib.pyplot as plt
plt.scatter(X, Y, color = 'black')
plt.xlabel('Anxiety disorders')
plt.ylabel('Depressive disorders')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#c) grafica los puntos y la recta de regresin
import statsmodels.api as sm
X_constante = sm.add_constant(X)
modelo = sm.OLS(Y, X_constante).fit()

b0, b1 = modelo.params

Fun = lambda X: b0 + b1 * X

Yc = Fun(X)

plt.plot(X, Yc, color = 'red', linestyle = '--')

#d) Coeficiente de correlacin y determinacin: Realiza los clculos necesarios
#para obtener la r de Pearson y el coeficiente de determinacin. Incluye dichos clculos
from scipy.stats import pearsonr
r,_ = pearsonr(X, Y)
print(f'coeficiente de correlacion: {r: 0.4f}\n')
print(f'coeficiente de determinacion: {r ** 2: 0.4f}\n')

#e) Interpreta los resultados del coeficiente de correlacin y el coeficiente de determinacin.
#El coeficiente de correlacin Indica una correlacin positiva muy dbil entre la prevalencia de depresin y ansiedad.
#El coeficiente de determinacin Esto significa que solo el 1.31% de la variabilidad en la prevalencia de ansiedad es explicada por la prevalencia de depresin.

#f) Calcula el intervalo de confianza del 95% para _1 y _0 Qu dice el intervalo de confianza a la posibilidad de que _1 sea igual a cero?
nivel_de_confianza = 0.95
intervalo_de_confianza = modelo.conf_int(alpha = 1 - nivel_de_confianza)
intervalo_de_confianza_b1 = intervalo_de_confianza.iloc[1]
print(f'intervalo de confianza para b1 de {nivel_de_confianza: 0.0%}')
print(f'{intervalo_de_confianza_b1[0]: 0.4f} < b1 < {intervalo_de_confianza_b1[1]: 0.4f}')
#Como este intervalo no incluye el valor 0, podemos concluir que hay evidencia estadsticamente significativa de que la pendiente _1 no es igual a 0.

#g) Realiza el grfico de los residuales: A partir de este grfico, menciona si los datos cumplen
#con los supuestos para la regresin: linealidad, normalidad en torno a la recta, homoscedasticidad
residuales = modelo.resid
plt.figure()
plt.scatter(X, residuales, color = 'black')
plt.xlabel('Anxiety disorders')
plt.ylabel('Depressive disorders')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.axhline(y = 0, color = 'red', linestyle = '--')

from scipy.stats import shapiro
_,valor_p_sh = shapiro(residuales)
print(f"valor_p de shepiro:{valor_p_sh}")

from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, X_constante)
print(f'Valor_p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')
#Se cumple la linealidad: Los residuales se distribuyen de manera relativamente aleatoria alrededor de la lnea roja,
#lo que indica que la relacin entre ansiedad y depresin es aproximadamente lineal.
#Normalidad de los residuales es dudosa: Hay cierta variabilidad en los extremos, lo que sugiere que podra haber desviaciones de la normalidad.
#No se cumple la homoscedasticidad: La dispersin de los residuales no es constante en toda la grfica, lo que indica heteroscedasticidad.
```
