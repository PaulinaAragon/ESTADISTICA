```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt


X = df[['edad',	'valor_acido_urico', 'valor_trigliceridos',  'valor_creatina']]
Y = df['masa_corporal']

# Datos de entrenamiento y datos de prueba

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()


Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de correlacion: {r2: 0.4f}\n')
```

```python
b0, b1, b2, b3, b4 = modelo.params

Fun = lambda x1, x2, x3, x4: b0 + b1 * x1 + b2 * x2 + b3 * x3 + b4 * x4

#si se trata de un modelo de predicción de demanda, un modelo de tiempo de espera
Fun(432, 60, 2, 120000)
```

```python
from statsmodels.formula.api import ols

moedo_2 = ols(formula = "masa_corporal ~ edad + valor_acido_urico + valor_trigliceridos + valor_creatina", data = df).fit()
tabla_anova = sm.stats.anova_lm(moedo_2, typ = 2)
tabla_anova
```

```python
# b) Realiza un gráfico con la dispersión y la recta de regresión ajustada.
import matplotlib.pyplot as plt

# Choose one feature from X to plot against Y
feature_to_plot = 'edad'  # For example, plotting 'edad' against 'masa_corporal'

plt.scatter(X[feature_to_plot], Y, color='darkviolet')
plt.xlabel(feature_to_plot)
plt.ylabel('masa_corporal')
plt.title('Scatter Plot')
plt.show()
```

```python
# C) Calcula el coeficiente de correlación y el coeficiente de determinación e interpreta los resultados.
import pandas as pd
from scipy.stats import pearsonr

# Calculate correlation for each feature in X with Y
correlation_results = {}
for feature in X.columns:
    r, _ = pearsonr(X[feature], Y)
    correlation_results[feature] = r

# Print correlation results
for feature, r in correlation_results.items():
    print(f'Coeficiente de correlación para {feature}: {r: 0.4f}')
    print(f'Coeficiente de determinación para {feature}: {r ** 2: 0.4f}\n')
```

```python
# g) Realiza el gráfico de los residuales: A partir de este gráfico, menciona si los datos cumplen con los supuestos para la regresión: linealidad, normalidad en torno a la recta, homoscedasticidad (hay más, pero con estas nos bastan).
residuales = modelo.resid
plt.figure()
# Instead of plotting X directly, plot the predicted values (Yc) against the residuals
plt.scatter(modelo.predict(X_train_constante), residuales, color = 'darkviolet')  # Changed X to modelo.predict(X_train_constante)
plt.xlabel('Valores Predichos')  # Changed x-axis label to reflect predicted values
plt.ylabel('Residuales')
plt.title('Gráfico de Residuales')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.axhline(0, color = 'red', linestyle = '--')
plt.show()
```
