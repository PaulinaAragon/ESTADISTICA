# **Metodolog√≠a**

*Introducci√≥n*

En este trabajo se analiza si los a√±os de experiencia influyen en el salario usando un modelo matem√°tico llamado regresi√≥n lineal. Se hacen c√°lculos para ver qu√© tan fuerte es la relaci√≥n entre estas dos cosas y si realmente est√°n conectadas. Adem√°s, se revisa si el modelo cumple ciertas reglas para asegurarnos de que los resultados sean confiables.

*Planteamiento de Hip√≥tesis*

Para evaluar la relaci√≥n entre los a√±os de experiencia y el salario, se establecen las siguientes hip√≥tesis:

Hip√≥tesis para el Modelo de Regresi√≥n Lineal
Hip√≥tesis Nula (ùêª0): No existe una relaci√≥n significativa entre los a√±os de experiencia y el salario, es decir, la pendiente de la regresi√≥n es igual a cero (ùõΩ1=0).

Hip√≥tesis Alternativa (ùêª1): Existe una relaci√≥n significativa entre los a√±os de experiencia y el salario, es decir, la pendiente de la regresi√≥n es distinta de cero (ùõΩ1‚â†0).

Hip√≥tesis para la Prueba de Normalidad (Shapiro-Wilk)

Hip√≥tesis Nula (ùêª0): Los residuos del modelo siguen una distribuci√≥n normal.

Hip√≥tesis Alternativa (ùêª1): Los residuos del modelo no siguen una distribuci√≥n normal.

Hip√≥tesis para la Prueba de Homocedasticidad (Breusch-Pagan)

Hip√≥tesis Nula (ùêª0): La varianza de los residuos es constante (homocedasticidad).

Hip√≥tesis Alternativa (ùêª1): La varianza de los residuos no es constante (heterocedasticidad).

*Tipo de Estudio*

Este estudio es de tipo cuantitativo y exploratorio, ya que busca analizar la relaci√≥n entre la experiencia laboral y el salario mediante herramientas estad√≠sticas.

*Marco Teorico*

La regresi√≥n lineal es un m√©todo estad√≠stico utilizado para modelar la relaci√≥n entre una variable dependiente ùëå y una variable independiente ùëã. En este caso, se analiza c√≥mo los a√±os de experiencia afectan el salario.

El modelo de regresi√≥n lineal simple se expresa como: ùëå=ùõΩ0 + ùõΩ1ùëã + ùúÄ

Donde:

Y es el salario.

X representa los a√±os de experiencia.

ùõΩ0 es la ordenada al origen.

ùõΩ1 es la pendiente, que indica el cambio esperado en el salario por cada a√±o adicional de experiencia.

Œµ es el error aleatorio.

Para evaluar la calidad del modelo, se calculan el coeficiente de correlaci√≥n y el coeficiente de determinaci√≥n, que indican qu√© tan bien la variable independiente explica la variaci√≥n en la variable dependiente. Adem√°s, se realizan pruebas estad√≠sticas para validar los supuestos del modelo, como normalidad de los residuos y homocedasticidad.

*Metogologia*

Obtenci√≥n de Datos:

Se utiliz√≥ un conjunto de datos que relaciona los a√±os de experiencia con el salario. Se eliminaron registros con valores faltantes.


An√°lisis Exploratorio:

Se realiz√≥ un gr√°fico de dispersi√≥n para visualizar la relaci√≥n entre ambas variables.

Ajuste del Modelo:

Se aplic√≥ una regresi√≥n lineal simple para encontrar la ecuaci√≥n que mejor representa la relaci√≥n entre experiencia y salario.

Coeficiente de correlaci√≥n:

Mide la fuerza y direcci√≥n de la relaci√≥n entre la experiencia y el salario.

Coeficiente de determinaci√≥n:

Indica el porcentaje de variabilidad en el salario que puede explicarse por los a√±os de experiencia.

Evaluacion del modelo:

Se calcularon los coeficientes de correlaci√≥n y determinaci√≥n para medir la fuerza de la relaci√≥n.


Se obtuvo un intervalo de confianza del 95% para la pendiente.


Se realiz√≥ un an√°lisis de varianza (ANOVA) para evaluar la significancia del modelo.

Verificacion de Supuestos:

Se aplic√≥ la prueba de Shapiro-Wilk para evaluar la normalidad de los residuos.

Se utiliz√≥ la prueba de Breusch-Pagan para verificar la homocedasticidad.

*Resultados*

Gr√°fico de dispersi√≥n: Se observ√≥ una relaci√≥n positiva entre los a√±os de experiencia y el salario.

Ecuaci√≥n de regresi√≥n: La ecuaci√≥n ajustada fue ùëå=ùëè0+ùëè1X, donde ùëè0 y ùëè1 fueron determinados a partir de los datos.

Coeficiente de correlaci√≥n: Se obtuvo un valor que indica una asociaci√≥n fuerte entre las variables.

Coeficiente de determinaci√≥n: El modelo explica un alto porcentaje de la variabilidad del salario.

Prueba de hip√≥tesis: Se rechaz√≥ la hip√≥tesis nula, confirmando que la experiencia influye significativamente en el salario.


Validaci√≥n del modelo:

La prueba de normalidad de Shapiro-Wilk mostr√≥ que los residuos siguen una distribuci√≥n aproximadamente normal.

La prueba de Breusch-Pagan indic√≥ que la varianza de los residuos es constante, validando la homocedasticidad del modelo.

*Conclusi√≥n*

Los resultados muestran que, en general, tener m√°s a√±os de experiencia s√≠ ayuda a ganar un mejor salario. El modelo confirma que hay una relaci√≥n entre ambas cosas y que los c√°lculos son correctos. Sin embargo, hay otros factores, como la educaci√≥n y el tipo de trabajo, que tambi√©n pueden influir en el salario, por lo que ser√≠a bueno analizarlos en estudios futuros.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/PaulinaAragon/ESTADISTICA/refs/heads/main/Salary_Data.csv')
# eliminar registros  cpn valores faltantes
df.dropna(inplace=True)

X = df["YearsExperience"] # Variable independiente
Y = df["Salary"] # Variable dependiente

# b) Realiza un gr√°fico con la dispersi√≥n y la recta de regresi√≥n ajustada.
import matplotlib.pyplot as plt
plt.scatter(X, Y, color = 'black')
plt.xlabel('YearsExperience') #nombrar eje x
plt.ylabel('Salary') #nombrar eje y
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# Recta de regresion lineal
import statsmodels.api as sm
X_constante = sm.add_constant(X)
modelo = sm.OLS(Y, X_constante).fit()
b0, b1 = modelo.params
Fun = lambda X: b0 + b1 * X
Yc = Fun(X)
plt.plot(X, Yc, color = 'red', linestyle = '--')

# c) Calcula el coeficiente de correlaci√≥n y el coeficiente de determinaci√≥n e interpreta los resultados.
from scipy.stats import pearsonr
r,_ = pearsonr(X, Y)
print(f'coeficiente de correlacion: {r: 0.4f}\n')
print(f'coeficiente de determinacion: {r ** 2: 0.4f}\n')

# d) Obt√©n un intervalo de confianza de 95% para la pendiente e interpreta el resultado. Respalda tu conclusi√≥n usando ANOVA.
nivel_de_confianza = 0.95
intervalo_de_confianza = modelo.conf_int(alpha = 1 - nivel_de_confianza)
intervalo_de_confianza_b1 = intervalo_de_confianza.iloc[1]
print(f'intervalo de confianza para b1 de {nivel_de_confianza: 0.0%}')
print(f'{intervalo_de_confianza_b1[0]: 0.4f} < b1 < {intervalo_de_confianza_b1[1]: 0.4f}')

# Tabla ANOVA
from statsmodels.formula.api import ols
# Y ~ X
modelo_2 = ols('YearsExperience ~ Salary', data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_2)
print(tabla_anova)

# e) Verifica los supuestos.
residuales = modelo.resid
plt.figure()
plt.scatter(X, residuales, color = 'black')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.axhline(y = 0, color = 'red', linestyle = '--')

from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f' valor_p Test de shapiro: {valor_p_sh: 0.4f}\n')

from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(residuales, X_constante)
print(f'Valor_p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')
```
